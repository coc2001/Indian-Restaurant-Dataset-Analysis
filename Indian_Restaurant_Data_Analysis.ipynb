# **6. Indian Restaurant Analysis by Format - Fastfood/Bakery/Street Food**

(i) Using Pie-Plot

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

filepath='after_drop_indian_restaurants.csv'
customerDf = pd.read_csv(filepath)
customerDf.dropna(inplace=True)
customerDf.drop_duplicates(inplace=True)
customerDf['fast_food_or_not'].value_counts().plot(kind='pie',autopct='%1.2f%%',colors=["#00FFFF","blue"])
plt.title('Restaurant Count - Fast Food or not')
plt.show()
customerDf['bakery_or_not'].value_counts().plot(kind='pie',autopct='%1.2f%%',colors=["#F88379","#FF4433"])
plt.title('Restaurant Count - Bakery or not')
plt.show()
customerDf['street_food'].value_counts().plot(kind='pie',autopct='%1.2f%%',colors=["#AFE1AF","#00A36C"])
plt.title('Restaurant Count - Street Food or not')
plt.show()
```

(ii) Using Bar-Diagram

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('after_drop_indian_restaurants.csv')

#for fast food
fast_food_counts = df['fast_food_or_not'].value_counts()
plt.bar(fast_food_counts.index.astype(str), fast_food_counts.values, color=['pink', 'red'])
plt.title('Fast Food Availability in Restaurants')
plt.xlabel('Fast Food Served (1) / Not Served (0)')
plt.ylabel('Restaurant Count')
plt.show()

#for bakery item
bakery_counts = df['bakery_or_not'].value_counts()
plt.bar(bakery_counts.index.astype(str), bakery_counts.values, color=['cyan', 'green'])
plt.title('Availability of Bakery Item in Restaurants')
plt.xlabel('Bakery Item Served (1) / Not Served (0)')
plt.ylabel('Restaurant Count')
plt.show()

#for street_food
street_food_counts = df['street_food'].value_counts()
plt.bar(street_food_counts.index.astype(str), street_food_counts.values, color=['yellow','blue'])
plt.title('Availability of Street Food in Restaurants')
plt.xlabel('Street Food Served (1) / Not Served (0)')
plt.ylabel('Restaurant Count')
plt.show()
```

# **7. Indian Restaurant Analysis by Cuisine - South-Indian/North-Indian/Biriyani**

(i) Using Pie-Plot

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

filepath='after_drop_indian_restaurants.csv'
customerDf = pd.read_csv(filepath)
customerDf.dropna(inplace=True)
customerDf.drop_duplicates(inplace=True)
customerDf['south_indian_or_not'].value_counts().plot(kind='pie',autopct='%1.2f%%',colors=["#FFBF00","#89CFF0"])
plt.title('Restaurant Count - South Indian or not')
plt.show()
customerDf['north_indian_or_not'].value_counts().plot(kind='pie',autopct='%1.2f%%',colors=["#CF9FFF","#BF40BF"])
plt.title('Restaurant Count - North Indian or not')
plt.show()
customerDf['biryani_or_not'].value_counts().plot(kind='pie',autopct='%1.2f%%',colors=["#DAA06D","brown"])
plt.title('Restaurant Count - Biryani or not')
plt.show()
```

(ii) Using Bar-Diagram

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('after_drop_indian_restaurants.csv')

#for south indian
south_indian_counts = df['south_indian_or_not'].value_counts()
plt.bar(south_indian_counts.index.astype(str), south_indian_counts.values, color=["#CF9FFF","#BF40BF"])
plt.title('South Indian Food Availability in Restaurants')
plt.xlabel('South-Indian Food Served (1) / Not Served (0)')
plt.ylabel('Restaurant Count')
plt.show()

#for north indian
north_indian_counts = df['north_indian_or_not'].value_counts()
plt.bar(north_indian_counts.index.astype(str), north_indian_counts.values, color=['yellow','red'])
plt.title('North Indian Food Availability in Restaurants')
plt.xlabel('North-Indian Food Served (1) / Not Served (0)')
plt.ylabel('Restaurant Count')
plt.show()

#for biryani
biryani_counts = df['biryani_or_not'].value_counts()
plt.bar(biryani_counts.index.astype(str), biryani_counts.values, color=['blue', 'orange'])
plt.title('Biryani Availability in Restaurants')
plt.xlabel('Biryani Served (1) / Not Served (0)')
plt.ylabel('Restaurant Count')
plt.show()
```

# **8. Analysing Average Price, Rating and Average Delivery Time of the Restaurants using Histogram**

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('after_drop_indian_restaurants.csv')

#for Average price
column_name = 'average_price'
marks_group = np.array([0,100,200,300,400,500,600,700,800])
plt.hist(df[column_name], bins=marks_group, color='cyan', edgecolor='blue')
plt.title('Histogram of {}'.format(column_name))
plt.xlabel(column_name)
plt.show()

#for rating
column_name = 'rating'
marks_group = np.array([2.0,2.2,2.4,2.6,2.8,3.0,3.2,3.4,3.6,3.8,4.0,4.2,4.4,4.6,4.8,5.0])
plt.hist(df[column_name], bins=marks_group, color='yellow', edgecolor='red')
plt.title('Histogram of {}'.format(column_name))
plt.xlabel(column_name)
plt.show()

#for average_delivery_time
column_name = 'average _delivery_time'
marks_group = np.array([15,20,25,30,35,40,45,50,55,60,65,70,75,80])
plt.hist(df[column_name], bins=marks_group, color='#AFE1AF', edgecolor='green')
plt.title('Histogram of {}'.format(column_name))
plt.xlabel(column_name)
plt.show()
```

# **9. Analysing the Number of  Restaurents According to Rating, Average Price and Average Delivery Time using Bar-Diagram**

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('after_drop_indian_restaurants.csv')
x_column = 'rating'
counts = df.groupby(x_column).size()
my_color=['purple','blue','skyblue','green','yellow','orange','red']
counts.plot(kind='bar', color=my_color, edgecolor='black')
plt.title('Rating-wise number of Restaurants')
plt.xlabel('rating')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('after_drop_indian_restaurants.csv')
x_column = 'average_price'
counts = df.groupby(x_column).size()
my_color=['red','cyan','blue','yellow','orange','pink','green']
counts.plot(kind='bar', color=my_color, edgecolor='black')
plt.title('Average Prise Wise number of Restaurants')
plt.xlabel('average_price')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('after_drop_indian_restaurants.csv')
counts = df.groupby('average _delivery_time').size()
counts.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Rating-wise number of Restaurants')
plt.xlabel('average_delivery_time')
plt.show()
```
